```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
iris=datasets.load_iris()
df=pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target']=iris.target
df['target_names']=df['target'].map({0:'setosa', 1:'versicolor', 2:'virginica'})

print(df.head())

```

```python
X=df.iloc[:,0:4]
y=df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
dt_model=DecisionTreeClassifier(
    criterion='gini',
    max_depth=4,
    min_samples_split=2,
    random_state=42
)
```

```python
dt_model.fit(X_train,y_train)
```

```python
y_pred=dt_model.predict(X_test)
```

```python
print("Accuracy (Gini index):", accuracy_score(y_test, y_pred))
plt.figure(figsize=(12,8))
plot_tree(dt_model,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
plt.show()
```

```python
dt_model2=DecisionTreeClassifier(
    criterion='entropy',
    max_depth=4,
    min_samples_split=2,
    random_state=42
)
dt_model2.fit(X_train,y_train)
```

```python
y_pred2=dt_model2.predict(X_test)
```

```python
print("Accoracy(Entropy):",accuracy_score(y_test,y_pred2))
plt.figure(figsize=(12,8))
plot_tree(dt_model2,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
plt.show()
```

```python
dt_model3=DecisionTreeClassifier(
    criterion='log_loss',
    max_depth=4,
    min_samples_split=2,
    random_state=42
)
dt_model3.fit(X_train,y_train)
```

```python
ypred3=dt_model3.predict(X_test)
```

```python
print("Accuracy(Log Loss):",accuracy_score(y_test,ypred3))
plt.figure(figsize=(12,8))
plot_tree(dt_model3,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
plt.show()
```
