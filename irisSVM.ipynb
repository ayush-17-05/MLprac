```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, precision_score , recall_score , accuracy_score, f1_score
```

```python
# Load Iris dataset
iris = datasets.load_iris()

# Create DataFrame for better visualization
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.to_csv('iris_dataset.csv', index=False)
df['target'] = iris.target
df['target_name'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

print(df.head())
print("\nData Information:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Linear KERNEL

```python
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
precision = precision_score(y_test, y_pred, average=None)
print("\nPrecision (per class):", precision)

# For overall precision
precision_macro = precision_score(y_test, y_pred, average='macro')
print("Average Precision (Macro):", precision_macro)

# Detailed report
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=iris.target_names))
```

```python
precision = precision_score(y_test, y_pred, average=None)
print("\nPrecision (per class):", precision)
recall = recall_score(y_test, y_pred, average=None)
print("\nRecall (per class):", recall)

# For overall precision and recall
print("Average Precision (Macro):", precision_score(y_test, y_pred, average='macro'))
print("Average Recall (Macro):", recall_score(y_test, y_pred, average='macro'))

accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy:", accuracy)

f1_measure = f1_score(y_test, y_pred, average=None)
print("\nF1-Measure (per class):", f1_measure)

print("Average F1-Measure (Macro):", f1_score(y_test, y_pred, average='macro'))

# Detailed report
# print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=iris.target_names))
```
